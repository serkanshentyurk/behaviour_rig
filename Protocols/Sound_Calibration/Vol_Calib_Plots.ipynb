```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
rig_volumes = {
    'Rig 1': [41.2, 44, 48, 55.5, 62.4, 69.2, 75.8, 82.5, 89.6, 89.8, 89.4],
    'Rig 2': [np.nan, 44, 45.4, 50.3, 56.8, 63.4, 70.4, 77.3, 82.5, 83.5, 83],
    'Rig 3': [np.nan, 44.2, 47.4, 53.2, 59.7, 66.6, 73.3, 80.2, 85.5, 86.6, 87.1],
    'Rig 4': [105, 115, 100, 90, 95, 80, 110, 120, 85, 100],
    'Rig 5': [120, 95, 105, 100, 90, 80, 110, 115, 85, 100],
    'Rig 6': [100, 110, 95, 105, 85, 90, 115, 120, 80, 100],
    'Rig 7': [105, 90, 100, 85, 95, 80, 110, 120, 100, 115],
    'Rig 8': [41.5, 45.9, 51.8, 58.9, 65.3, 72, 78.4, 84.9, 91.8, 98.6, 102.5]
}

bonsai_sound_levels = np.arange(50,105,5)
```

```python
# plot rig volumes Vs bonsai sound levels
plt.figure(figsize=(6,6))
plt.plot(bonsai_sound_levels, rig_volumes['Rig 1'], 'o-', label='Rig 1')
plt.plot(bonsai_sound_levels, rig_volumes['Rig 2'], 'o-', label='Rig 2')
plt.plot(bonsai_sound_levels, rig_volumes['Rig 3'], 'o-', label='Rig 3')
plt.plot(bonsai_sound_levels, rig_volumes['Rig 8'], 'o-', label='Rig 8')
# add horizontal lines at 50 and 82
plt.axhline(50, color='k', linestyle='--')
plt.axhline(82, color='k', linestyle='--')
plt.xlabel('Bonsai Sound Level')
plt.ylabel('Rig Volume')
plt.legend()

plt.show()
```

```python
# plot rig volumes Vs bonsai sound levels, but only for the values that are [3:8]
plt.figure(figsize=(6,6))
plt.plot(bonsai_sound_levels[2:9], rig_volumes['Rig 1'][2:9], 'o-', label='Rig 1')
plt.plot(bonsai_sound_levels[2:9], rig_volumes['Rig 2'][2:9], 'o-', label='Rig 2')
plt.plot(bonsai_sound_levels[2:9], rig_volumes['Rig 3'][2:9], 'o-', label='Rig 3')
plt.plot(bonsai_sound_levels[2:9], rig_volumes['Rig 8'][2:9], 'o-', label='Rig 8')
plt.xlabel('Bonsai Sound Level')
plt.ylabel('Rig Volume')

# fit a line to the data for each rig
from scipy.stats import linregress
slope, intercept, r_value, p_value, std_err = linregress(bonsai_sound_levels[3:8], rig_volumes['Rig 1'][3:8])
print('Rig 1: slope = {}, intercept = {}, r_value = {}, p_value = {}, std_err = {}'.format(slope, intercept, r_value, p_value, std_err))
plt.plot(bonsai_sound_levels[2:9], intercept + slope*bonsai_sound_levels[2:9], 'k--')

slope, intercept, r_value, p_value, std_err = linregress(bonsai_sound_levels[3:8], rig_volumes['Rig 2'][3:8])
print('Rig 2: slope = {}, intercept = {}, r_value = {}, p_value = {}, std_err = {}'.format(slope, intercept, r_value, p_value, std_err))
plt.plot(bonsai_sound_levels[2:9], intercept + slope*bonsai_sound_levels[2:9], 'k--')

slope, intercept, r_value, p_value, std_err = linregress(bonsai_sound_levels[3:8], rig_volumes['Rig 3'][3:8])
print('Rig 3: slope = {}, intercept = {}, r_value = {}, p_value = {}, std_err = {}'.format(slope, intercept, r_value, p_value, std_err))
plt.plot(bonsai_sound_levels[2:9], intercept + slope*bonsai_sound_levels[2:9], 'k--')

slope, intercept, r_value, p_value, std_err = linregress(bonsai_sound_levels[3:8], rig_volumes['Rig 8'][3:8])
print('Rig 8: slope = {}, intercept = {}, r_value = {}, p_value = {}, std_err = {}'.format(slope, intercept, r_value, p_value, std_err))
plt.plot(bonsai_sound_levels[2:9], intercept + slope*bonsai_sound_levels[2:9], 'k--')


plt.legend()
plt.show()
```
