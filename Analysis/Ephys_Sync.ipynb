# Test With Harp Clock

```python
df_receive_event_data_32 = pd.read_csv(ephys_path + '/Pressure_Valve_Receive2023-12-06T20_33_37.9960320+00_00/event_data_32.csv')
df_receive_event_data_44 = pd.read_csv(ephys_path + '/Pressure_Valve_Receive2023-12-06T20_33_37.9960320+00_00/event_data_44.csv')
df_receive_write_data = pd.read_csv(ephys_path + '/Pressure_Valve_Receive2023-12-06T20_33_37.9960320+00_00/write_data.csv')

df_send_event_data_44 = pd.read_csv(ephys_path + '/Water_Valve_Send2023-12-06T20_33_38.0021248+00_00/event_data_44.csv')
df_send_write_data = pd.read_csv(ephys_path + '/Water_Valve_Send2023-12-06T20_33_38.0021248+00_00/write_data.csv')
# only keep rows of df_send_write_data where PORT2_DO not nan
df_send_write_data = df_send_write_data[df_send_write_data['PORT2_DO'].notna()]
```

```python
raise Exception('stop')
```

```python
# plot Payload Vs Timestamp for each event_data_32
fig, ax  = plt.subplots(1, 1, figsize=(20, 5))
plt.plot(df_receive_event_data_32['Timestamp'], df_receive_event_data_32['Payload']*0.2, 'o', label = 'Receive', alpha = 0.5)
plt.plot(df_send_write_data['Times'], df_send_write_data['PORT2_DO'], 'o', label = 'Send', alpha = 0.5)

plt.xlabel('Time')
plt.legend()
plt.show()
```

```python
# take dfference between df_send_write_data.Times.values and df_receive_event_data_32.Timestamp.values
diff_vec = df_send_write_data.Times.values - df_receive_event_data_32.Timestamp.values
```

```python
# make a histogram of the time difference
fig, ax  = plt.subplots(1, 1, figsize=(5, 5))

plt.hist(diff_vec, bins = 100)
plt.xlabel('Time Difference')
plt.show()
```

```python
# plot diff_vec

fig, ax  = plt.subplots(1, 1, figsize=(20, 5))
plt.plot(diff_vec, 'o', alpha = 0.5)
plt.xlabel('Time Difference')
plt.show()
```
